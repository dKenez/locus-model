```python
import json
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 12

# Read the JSON file
with open('results.json') as file:
    data = json.load(file)

# Read the JSON file
with open('results_db.json') as file:
    data_db = json.load(file)

# Treat the data as a graph
# Your code for working with the graph goes here
```

```python
x = data["X"]
y = data["Y_mean"]
per_element_time = data["Y_avg_time"]
```

```python
# Create figure and axis objects
fig, ax1 = plt.subplots()

# Plot y_mean as a bar chart
ax1.plot(x, y, color='#1F3DFF', marker='.')
ax1.set_xlabel('Batch size [1]')
ax1.set_ylabel('Batch fetch time [s]', color='#1F3DFF')
ax1.tick_params(axis='y', labelcolor='#1F3DFF')

# Create a second y-axis
ax2 = ax1.twinx()

# Plot per_element_time as a line chart
ax2.plot(x, per_element_time, color='#C4000D', marker='.')
ax2.set_ylabel('Record fetch time [s]', color='#C4000D')
ax2.tick_params(axis='y', labelcolor='#C4000D')

# Set the x-axis to log scale
ax1.set_xscale('log')

# Set the title and show the plot
plt.title('Data storage performance with .parquet files')
plt.savefig("parquet_speeds.svg", bbox_inches='tight', transparent=True,
               pad_inches=0, format="svg")
plt.show()
```
